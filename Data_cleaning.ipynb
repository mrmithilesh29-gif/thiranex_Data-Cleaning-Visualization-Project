# Feature Engineering

### Let us extract some important features from existing features

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()       # We can see that the columns "Date" , "Month" and "Year" are added.
```

### Now we can remove the "Date_of_Journey" column.

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Year'].unique()
```

### As year column has only one value "2019". It is better to drop this column

```python
df.drop("Year",axis=1,inplace=True)
```

```python
df.head()
```

### We can see that the Arrival Time contains date and month also so we have to remove them and Only store the Arrival time

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(" ").str[0]
```

```python
df.head()
```

### Now split the Arrival Time and Departure Time into ArrivalTime_Hours, ArrivalTime_Hours, DepTime_Hours, DepTime_Min.

```python
df['ArrivalTime_Hours'] = df['Arrival_Time'].str.split(':').str[0]
df['ArrivalTime_Min'] = df['Arrival_Time'].str.split(':').str[1]
df['DepTime_Hours'] = df['Dep_Time'].str.split(':').str[0]
df['DepTime_Min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.head()
```

### Now we can drop the Arrival_Time and Dep_Time

```python
df.drop(['Dep_Time','Arrival_Time'],axis=1,inplace=True)
```

```python
df.head()
```

### The route is also not required so that is not a important column

```python
df.drop("Route",axis=1,inplace=True)
```

```python
df.head()
```

### Coverting the Duration into minutes

```python
df['Duration'].unique()
```

```python
def func(x):
    h=0
    m=0
    part = 0
    if 'h' in x:
        h=int(x.split('h')[0])
    
    if 'm' in x:
        part = x.split('h')[-1].replace('m','').strip()

    if part :
        m=int(part)

    return h*60 + m
df['Duration'] = df['Duration'].apply(func)
```

```python
df.head()
```

### Now let us the fill the values of Total Stops in numbers(0,1,2,3,4)

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].replace({
    np.nan : 1,
    "1 stop" :  1,
    "non-stop": 0,
    "2 stops" : 2,
    "3 stops" : 3,
    "4 stops" : 4
},inplace=True)
```

```python
df.head()
```

```python
df.info()
```

###  Now convert the columns into int type

```python
df['Total_Stops'] = df['Total_Stops'].astype(int)
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['ArrivalTime_Min'] = df['ArrivalTime_Min'].astype(int)
df['DepTime_Min'] = df['DepTime_Min'].astype(int)
df['DepTime_Hours'] = df['DepTime_Hours'].astype(int)
df['ArrivalTime_Hours'] = df['ArrivalTime_Hours'].astype(int)
```

### "Additional_Info" Column is also not much important

```python
df.drop('Additional_Info',axis=1,inplace=True)
```
