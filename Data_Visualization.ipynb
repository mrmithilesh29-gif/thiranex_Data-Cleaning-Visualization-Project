# Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("cleaned_dataset.csv")
```

## Now its time to Visualize the data and analyze them.

### Now let us see the Average Price cost of Different Airlines

```python
sns.barplot(x='Price' , y="Airline" , data=df,palette="deep")
plt.title("Airline Price Comparision")
```

### We can see that the Highest Average Flight price is Jet Airways Business. The lowest is Truget

```python
sns.boxplot(x='Price' , y = "Airline" , data = df,palette="muted")
plt.title("Outliers")
```

### We can observe many outliers in Jet Airways, Air India, and Indigo. And the Jet Airways Business has the longest outlier  ( aproximately 80000)

```python
sns.barplot(y="Price" , x="Total_Stops", data=df,palette="bright")
plt.title("Price Based on Total_Stops")
```

#### We can observe that as the Stops increases the price of the flight ticket alao increases.
## Flights with more Stops are more Costlier

```python
sns.scatterplot(x="Price" , y="Duration", data=df)
plt.title("Price Based on Duration")
```

### Most of the flights price's are below 25000 only and their are some outliers that are more than 30000

```python
sns.barplot(x='Source' , y="Price", data=df,palette='rainbow')
plt.title("Price based on Souce")
```

### The flights staring from Delhi are really of High price ranging upto 10000 and above.
### The flights Staring from Chennai are low compared to all the places.

### As Delhi is the capital of India flight price can be High

```python
sns.barplot(x='Destination' , y="Price", data=df,palette='pastel')
plt.title("Price based on Destination")
```

### The flights Whose destination is New Delhi are also having the highest fair compared to other places.
## This Clearly tells us that The Flights from or to Delhi are Costlier.

```python
airline_counts = df["Airline"].value_counts()
```

```python
plt.pie(x=airline_counts , labels= airline_counts.index,autopct='%1.1f%%')
plt.title("Percentage of Flights per Airline")
plt.tight_layout()
```

### We can observe that the Jet Airways has the Highest Number of flights and , Second Highest is Indigo.

```python
sns.barplot(y='Airline' , x ="Duration" , data = df,palette="dark")
plt.title("Flight Duration by Airline")
```

### Air India take the Highest Duration of Travelling , Second highest Duration is Jet Airways

```python
sns.boxplot(y='Airline' , x ="Total_Stops" , data = df,palette="colorblind")
plt.title("Distribution of Stops by Airline")
```

```python
sns.barplot(x="Total_Stops" , y="Duration" , data=df,palette="husl")
plt.title("Flight Duration by Number of Stops")
```

### As the number of Stops increases the Duration of the flight also increases

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True),cmap='coolwarm',annot=True)
```

### We can clearly observe the correlations between features using the Heatmap()
